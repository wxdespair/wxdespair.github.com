```python
import numpy as np
a = np.random.rand(100000, 3, 3)
print("随机生成的矩阵为：\n")
print(a)
print("\n矩阵的行列式值为：\n")
det_a = np.linalg.det(a)
print(det_a)
print("\n矩阵的逆为：\n")
inv_a = np.linalg.inv(a) 
print(inv_a)
```
