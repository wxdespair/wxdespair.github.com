## Matplotlib 可视化基础

Matplotlib 是 Python 语言和 Python 的数值数学扩展包的画图程序库，由各种可视化类构成，内部结构复杂。它提供了

* 一个面向对象的 API 接口，用于把画图嵌入到其它的应用程序中。
* 一个面向过程的 pylab 接口，和 MATLAB 的画图很类似。

Matplotlib 的作者是 John D. Hunter

* 屏幕(screen): 由很多像素(pixel)组成 
* 画布(figure): 容纳所有画形元素的最高层容器
* 坐标系(axes): 与数学中的坐标系相对应，由坐标原点、坐标轴组成。每个要展示的**数据点**会对应一个有序实数对，即坐标。
* 坐标轴(axis): 坐标系的一个数据维度
* 标签(label): 对图例进行说明的文字等内容。

plt.plot(x, y, format_string, **kwargs)

* x：X轴数据，列表或数组，可选
* y：Y轴数据，列表或数组
* format_string：控制曲线的格式字符串，可选
* \*\*kwargs：第二组或更多 (x, y, format_string)，当绘制多条曲线时，各条曲线的x不能省略

````python
import matplotlib.pyplot as plt
plt.plot([3, 4, 2, 5, 6])
plt.ylabel("grack")   # 指定y轴标识，还有 .xlabel()函数
# plt.savefig('plot_image_1', dpi=600)
# plt.savefig() 将输出图片存储为文件，默认PNG格式，
    # 可以通过dpi修改输出质量
plt.show()
````

````python
plt.plot([0, 2, 4, 6, 8], [3, 1, 4, 5, 2])
# plt.plot(x, y) 当有两个以上参数时，按照X轴和Y轴顺序绘制数据点
plt.ylabel("g")
plt.axis([-1, 10, 0, 6])
# plt.axis() 函数用来指定坐标轴的刻度跨值
plt.show
````

### 不同的画图环境

演示三种环境下的绘图

* 脚本

**注意：** `plt.show()` 命令的用法。

**画图流程：**
* 创建画布
* 创建坐标系，注意同一副画布上可以创建多个坐标系
* 在坐标系上绘图。

````python
import matplotlib.pyplot as plt
# 打开一个画布
plt.figure()

# 用户可以指定画布的大小
fig = plt.figure(figsize=(16,8), dpi=100) # DPI（Dots Per Inch，每英寸点数）


# 画布中添加一个坐标系
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 

# 画一幅函数图像, 显示或者保存
import numpy as np
x = np.linspace(0, 10, 10)
y = np.sin(x)
 
# 创建画布
fig = plt.figure()
# 创建坐标轴
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # 第一个坐标轴的范围
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # 第二个坐标轴的范围
 
axes1.plot(x, y, 'r')
axes2.plot(x, y, 'g')
plt.show() # 显示画布，无论画布上有无图像都可以显示
# fig.savefig('myfirstfig.png')
````

而在以ipython为核心的jupyter中可以方便的展示画布而省略plt.show()，只需在绘图单元格前添加魔术命令：

`%matplotlib inline`

即可默认显示画布。

````python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(0, 10, 1000)
y = np.sin(x)

# 创建画布
fig = plt.figure()

# 创建坐标系
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # 第一个坐标系的范围
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # 第二个坐标系的范围

axes1.plot(x, y, 'r')
axes2.plot(x, y, 'g')
````

### 子绘图划分

**方法其一：**

plt.subplot2grid(GridSpec, CurSpec, colspan=1, rowspan=1)

* 理念：设定网格，选中网格，确定选中行列区域数量，标号从0开始
* GridSpec：指定将区域如何划分
* CurSpec：指定选中区域在整个区域的位置
* colspan：同行方向的延伸
* rowspan：同列方向的延伸
* 例：plt.subplot2grid((3, 3), (1, 0), colspan=2)

````python
import matplotlib.pyplot as plt
import numpy as np
# 在相应语句下书写的语句就会在相应的区域绘制图片
plt.subplot2grid((3, 3), (0, 0), colspan=3)
plt.text(0.5, 0.5, "ax1",fontsize=10)
plt.subplot2grid((3, 3), (1, 0), colspan=2)
plt.text(0.5, 0.5, "ax2",fontsize=10)
plt.subplot2grid((3, 3), (1, 2), rowspan=2)
plt.text(0.5, 0.5, "ax3",fontsize=10)
plt.subplot2grid((3, 3), (2, 0))
plt.text(0.5, 0.5, "ax4",fontsize=10)
plt.subplot2grid((3, 3), (2, 1))
plt.text(0.5, 0.5, "ax5",fontsize=10)
plt.show()
````

**方法其二:**

pyplot 的绘图区域

`plt.subplot(nrows, ncols, plot_number)`
    
 nrows 为分区的行数，ncols代表分区的列数，plot_number代表当前图像在分区中的位置

````python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
a = np.arange(0.0, 5.0, 0.02)
plt.subplot(2, 1, 1)
plt.plot(a, f(a))
plt.subplot(2, 1, 2)
plt.plot(a, np.cos(2*np.pi*a), 'r--')
plt.show()
````

**方法其三：**

使用 GridSpec 类

````python
import matplotlib.gridspec as gridspec
gs = gridspec.GridSpec(3, 3)
# 可以使用类似二维切片的方式选取区域
ax1 = plt.subplot(gs[0, :])
plt.text(0.5, 0.5, "ax1",fontsize=10)
ax2 = plt.subplot(gs[1, :-1])
plt.text(0.5, 0.5, "ax2",fontsize=10)
ax3 = plt.subplot(gs[1:, -1])
plt.text(0.5, 0.5, "ax3",fontsize=10)
ax4 = plt.subplot(gs[2, 0])
plt.text(0.5, 0.5, "ax4",fontsize=10)
ax5 = plt.subplot(gs[2, 1])
plt.text(0.5, 0.5, "ax5",fontsize=10)
plt.show()
````

### 辅助文本信息的显示

| 函数           | 说明                            |
| :--------------: | :-------------------------------: |
| plt.xlabel()   | 对X轴增加文本标签               |
| plt.ylabel()   | 对Y轴增加文本标签               |
| plt.title()    | 对图形整体增加文本标签 （表名） |
| plt.text()     | 在任意位置增加文本              |
| plt.annotate() | 在图形中增加带箭头的注解        |

````python
import numpy as np
import matplotlib.pyplot as plt
a = np.arange(0.0, 5.0, 0.02)
plt.plot(a, np.cos(2*np.pi*a), 'r--')
plt.xlabel('横轴：时间', fontproperties='SimHei', fontsize=15, color='green')
plt.ylabel('纵轴：振幅', fontproperties='SimHei', fontsize=15)
plt.title(r'正弦波实例 $y=cos(2\pi x)$',fontproperties='SimHei', fontsize=25)
plt.text(2, 1.5, r'$\mu=100$',fontsize=15)
# 其中前两个参数为文本在图中的坐标位置
plt.axis([-1, 6, -2, 2])
plt.grid(True)  # 显示网格
plt.show()
````

### 极坐标（玫瑰花图）

````python
N = 20 # 图中图形的数量
theta = np.linspace(0.0, 2*np.pi, N, endpoint=False)
radii = 10*np.random.rand(N)
width = np.pi / 4 * np.random.rand(N)
ax = plt.subplot(111, projection='polar')
# 这里运用面相对象的理念绘制的极坐标图，而不是直接使用绘制极坐标图的函数
bars = ax.bar(theta, radii, width=width, bottom=0.0)
# 前三个参数分别对应极坐标中的left, height, width
for r, bar in zip(radii, bars):
    bar.set_facecolor(plt.cm.viridis(r/10.))
    bar.set_alpha(0.5)
plt.show() 
````

````python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates
from datetime import datetime

try:
    # Try to fetch a list of Matplotlib releases and their dates
    # from https://api.github.com/repos/matplotlib/matplotlib/releases
    import urllib.request
    import json

    url = 'https://api.github.com/repos/matplotlib/matplotlib/releases'
    url += '?per_page=100'
    data = json.loads(urllib.request.urlopen(url, timeout=.4).read().decode())

    dates = []
    names = []
    for item in data:
        if 'rc' not in item['tag_name'] and 'b' not in item['tag_name']:
            dates.append(item['published_at'].split("T")[0])
            names.append(item['tag_name'])
    # Convert date strings (e.g. 2014-10-18) to datetime
    dates = [datetime.strptime(d, "%Y-%m-%d") for d in dates]

except Exception:
    # In case the above fails, e.g. because of missing internet connection
    # use the following lists as fallback.
    names = ['v2.2.4', 'v3.0.3', 'v3.0.2', 'v3.0.1', 'v3.0.0', 'v2.2.3',
             'v2.2.2', 'v2.2.1', 'v2.2.0', 'v2.1.2', 'v2.1.1', 'v2.1.0',
             'v2.0.2', 'v2.0.1', 'v2.0.0', 'v1.5.3', 'v1.5.2', 'v1.5.1',
             'v1.5.0', 'v1.4.3', 'v1.4.2', 'v1.4.1', 'v1.4.0']

    dates = ['2019-02-26', '2019-02-26', '2018-11-10', '2018-11-10',
             '2018-09-18', '2018-08-10', '2018-03-17', '2018-03-16',
             '2018-03-06', '2018-01-18', '2017-12-10', '2017-10-07',
             '2017-05-10', '2017-05-02', '2017-01-17', '2016-09-09',
             '2016-07-03', '2016-01-10', '2015-10-29', '2015-02-16',
             '2014-10-26', '2014-10-18', '2014-08-26']

    # Convert date strings (e.g. 2014-10-18) to datetime
    dates = [datetime.strptime(d, "%Y-%m-%d") for d in dates]
# Choose some nice levels
levels = np.tile([-5, 5, -3, 3, -1, 1],
                 int(np.ceil(len(dates)/6)))[:len(dates)]

# Create figure and plot a stem plot with the date
fig, ax = plt.subplots(figsize=(8.8, 4), constrained_layout=True)
ax.set(title="Matplotlib release dates")

markerline, stemline, baseline = ax.stem(dates, levels,
                                         linefmt="C3-", basefmt="k-",
                                         use_line_collection=True)

plt.setp(markerline, mec="k", mfc="w", zorder=3)

# Shift the markers to the baseline by replacing the y-data by zeros.
markerline.set_ydata(np.zeros(len(dates)))

# annotate lines
vert = np.array(['top', 'bottom'])[(levels > 0).astype(int)]
for d, l, r, va in zip(dates, levels, names, vert):
    ax.annotate(r, xy=(d, l), xytext=(-3, np.sign(l)*3),
                textcoords="offset points", va=va, ha="right")

# format xaxis with 4 month intervals
ax.get_xaxis().set_major_locator(mdates.MonthLocator(interval=4))
ax.get_xaxis().set_major_formatter(mdates.DateFormatter("%b %Y"))
plt.setp(ax.get_xticklabels(), rotation=30, ha="right")

# remove y axis and spines
ax.get_yaxis().set_visible(False)
for spine in ["left", "top", "right"]:
    ax.spines[spine].set_visible(False)

ax.margins(y=0.1)
plt.show()
````

## 速查

1. 绘制折线图（plt.plot）
1. 设置图片大小和分辨率（plt.figure）
1. 保存图片到本地（plt.savefig）
1. 设置坐标轴上的刻度和字符串（xticks、yticks）
1. 设置坐标轴单位或提示和标题（lable(title,xlable,ylable)）
1. 指定字体（font_manager + fontProperties或matplotlib.rc）
1. 在单一图表上绘制多条线条（plt多次即可）

**fig = plt.figure(figsize = (20, 8), dpi = 72)**
* figure是图形图标的意思，在这里指产生的图
* 通过实例化一个figure并且传入参数，就可以在后台自动使用该figure实例
* 在图像模糊的时候可以传入dpi参数，让图片更加清晰

**plt.savefig("./sig_size.png")**
* 保存图片
* 还可以保存svg的矢量图格式，放大无锯齿

**plt.xticks(numpy.arange(2, 25, 0.5))**
* 设置x轴的刻度
* 只要输入的参数为一个可作为刻度的数据即可，列表或数组啥的都可以

**x = range(0,20)**<br>
**_x = x**<br>
**_xtick_labels = ["hello,{}".format(i) for i in _x]**<br>
**plt.xticks(x,_xtick_labels)**<br>
* 单设置一个列表，存放坐标轴要表示出的字符串刻度
* 上面的**x**与**_xtick_labels**要有相同的元素数

**plt.xticks(list(x)[::10],_xtick_labels[::10],rotation=90)**
* 前两个参数只要是相同长短的列表就可以
* rotation的值表示刻度旋转的角度

**from matplotlib import font_manager**<br>
**my_font = font_manager.FontProperties(fname = '一纸情书.ttf')**<br><br>
**plt.xticks(list(x)[::10],_xtick_labels[::10],rotation=30,fontproperties=my_font)**<br>
* 前两句是导入库和根据字体包创建自己在图中要显示的字体
* 第三句是在应用中文汉字的地方，指定使用自己上面创建的字体

**plt.grid(alpha = 0.4)**
* 为图标添加网格
* alpha参数控制网格透明度

**plt.plot(x, y_1,label="自己",color="red")**<br>
**plt.plot(x, y_2,label="同桌",color="black")**<br><br>
**plt.legend(prop = my_font, loc="upper left")**
* 前两句是必要条件，有两个因变量，
* label参数是为图例做准备
* color参数指定图表上线条的颜色
    * r <=> 红色
    * g <=> 绿色
    * b <=> 蓝色
    * w <=> 白色
    * c <=> 青色
    * a <=> 洋红
    * y <=> 黄色
    * k <=> 黑色
    * 或者直接写颜色的16进制色号
* linestyle参数指定线条风格
    *  **-** <=> 实线
    * --  <=> 虚线
    * -.  <=> 点划线
    * :   <=> 点虚线
* linewidth参数指定线条粗细
* color、linstyle、linewidth也都适用于网格的线条属性，同时alpha控制不透明度也适用于该线条
* 第三句为图表创建图例
* prop为图例指定文字字体，且只在此处，其余地方都用fontproperties=my_font指定
* loc为图例指定显示位置，默认位于图表右上方

**线条样式补充**

    '-'	实线样式
    '--'	短横线样式
    '-.'	点划线样式
    ':'	虚线样式
    '.'	点标记
    ','	像素标记
    'o'	圆标记
    'v'	倒三角标记
    '^'	正三角标记
    '&lt;'	左三角标记
    '&gt;'	右三角标记
    '1'	下箭头标记
    '2'	上箭头标记
    '3'	左箭头标记
    '4'	右箭头标记
    's'	正方形标记
    'p'	五边形标记
    '*'	星形标记
    'h'	六边形标记 1
    'H'	六边形标记 2
    '+'	加号标记
    'x'	X 标记
    'D'	菱形标记
    'd'	窄菱形标记
    '&#124;'	竖直线标记
    '_'	水平线标记

### 添加坐标系

````python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(0, 10, 1000)
y = np.sin(x)

# 创建画布
fig = plt.figure()

# 创建坐标系
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # 第一个坐标系的范围
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # 第二个坐标系的范围

axes1.plot(x, y, 'r')
axes2.plot(x, y, 'g')
````
